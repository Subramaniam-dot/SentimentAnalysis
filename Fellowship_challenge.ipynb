Function for Number of words in a text

```python
def no_of_words(text):
    words= text.split()
    word_count = len(words)
    return word_count
```

Data Preprocessing, to replace unnecessary characters to space

```python
#funcion to replicate the procedure each time
#re -> regular expression

def data_processing(text):
    text= text.lower()
    text = re.sub('<br />', '', text)
    text = re.sub(r"https\S+|www\S+|http\S+", '', text, flags = re.MULTILINE)
    text = re.sub(r'\@w+|\#', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text_tokens = word_tokenize(text)
    filtered_text = [w for w in text_tokens if not w in stop_words] 
    return " ".join(filtered_text)
```

```python
df.review = df['review'].apply(data_processing)
```

```python
X = df['review']
Y = df['sentiment']
vect = TfidfVectorizer()
X = vect.fit_transform(df['review'])
```

```python
import xgboost as xgb
import pandas as pd

# Load data into a Pandas dataframe
# df = pd.read_csv('data.csv')

# Split the data into features and labels
# X = df.drop(columns='label')
# y = df['label']

# Split the data into training and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

# Create the XGBClassifier object
xgbc = xgb.XGBClassifier(reg_alpha=0.1, reg_lambda=0.5, random_state=99)

# Fit the model to the training data
xgbc.fit(X_train, y_train)

# Make predictions on the test data
# predictions = xgbc.predict(X_test)


yt_p = xgbc.predict(X_train)
yv_p = xgbc.predict(X_test)

# Evaluate the model's performance
# from sklearn.metrics import accuracy_score
# accuracy = accuracy_score(y_test, predictions)
# print(f'Test accuracy: {accuracy:.2f}')
```

```python
logreg_acc = accuracy_score(yv_p, y_test)

print("Test accuracy: {:.2f}%".format(logreg_acc*100))
print(confusion_matrix(y_test, yv_p))
print("\n")
print(classification_report(y_test, yv_p))
```

These are evaluation metrics for a binary classification model. The metrics are calculated usings the results of the model's predictions on a test dataset.

precision is the number of true positive predictions made by the model, divided by the total number of positive predictions made by the model. It measures the proportion of positive predictions that were correct.

recall is the number of true positive predictions made by the model, divided by the total number of actual positive instances in the test dataset. It measures the proportion of actual positive instances that were correctly predicted by the model.

f1-score is the harmonic mean of precision and recall. It is a single metric that combines both precision and recall into a single score, with a value between 0 and 1.

support is the number of instances in the test dataset for each class. In this case, there are 4946 instances of class 0 and 4970 instances of class 1.

accuracy is the number of correct predictions made by the model, divided by the total number of predictions made. It measures the overall accuracy of the model.

macro avg is the average of the precision, recall, and f1-score for each class, calculated independently.

weighted avg is the average of the precision, recall, and f1-score for each class, weighted by the number of instances in each class. It gives more weight to the metrics for the class with more instances.

we can see from the metrics that the performace of the model is good
